```python
import numpy as np
import matplotlib.pyplot as plt
import time
import seaborn as sns

def create_data(n):
    return np.random.randint(10, size=n)

def linear(A):
    for i in range(len(A)):
        element = A[i] * 2


sizes = [10, 100, 1000, 10000, 100000, 1000000]
times = []
for i in sizes:
    A = create_data(i)
    start = time.time()
    linear(A)
    end = time.time()
    total = end - start
    times.append(total)
    print("N: ", i, " with time", total)

sns.set_theme()

plt.plot(sizes, times, c = "r", label = "Lineal")
plt.xlabel("N")
plt.ylabel("Time in seconds")
plt.title("Linear Function")
plt.legend()
plt.show()
```

```python
def quadratic(A):
    for i in range(len(A)):
        for j in range(len(A)):
            #print("Element: ", A[i])
            element = A[i] * 2

sizes = [10, 100, 1000, 10000]
times = []
for i in sizes:
    A = create_data(i)
    start = time.time()
    quadratic(A)
    end = time.time()
    total = end - start
    times.append(total)
    print("N: ", i, " with time", total)

sns.set_theme()

plt.plot(sizes, times)
plt.plot(sizes, times, c = "r", label = "Lineal")
plt.xlabel("N")
plt.ylabel("Time in seconds")
plt.title("Cuadratic Function")
plt.legend()
plt.show()
```

```python
def cubic(A):
    for i in range(len(A)):
        for j in range(len(A)):
            for k in range(len(A)):
                element = A[i] * 2

sizes = [10, 100, 1000]
times = []
for i in sizes:
    A = create_data(i)
    start = time.time()
    cubic(A)
    end = time.time()
    total = end - start
    times.append(total)
    print("N: ", i, " with time", total)

sns.set_theme()

plt.plot(sizes, times)
plt.plot(sizes, times, c = "r", label = "Lineal")
plt.xlabel("N")
plt.ylabel("Time in seconds")
plt.title("Cubic Function")
plt.legend()
plt.show()
```

```python
def combined(A):
    for i in range(len(A)): # O(n^3)
        for j in range(len(A)):
            for k in range(len(A)):
                element = A[i] * 2
    for i in range(len(A)): # O(n)
            element = A[i] * 2
    for j in range(len(A)): # O(n)
            element = A[i] * 2

sizes = [10, 100, 1000]
times = []
for i in sizes:
    A = create_data(i)
    start = time.time()
    combined(A)
    end = time.time()
    total = end - start
    times.append(total)
    print("N: ", i, " with time", total)

sns.set_theme()

plt.plot(sizes, times)
plt.plot(sizes, times, c = "r", label = "Lineal")
plt.xlabel("N")
plt.ylabel("Time in seconds")
plt.title("O(n^3)+O(2n) Function")
plt.legend()
plt.show()
```

In this graph, we can see the comparison between the execution times of the different functions
